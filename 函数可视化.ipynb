## 用Python的'matplotlib'库，把数学函数从**符号公式**变成**视觉图像**

```python
# numpy 的作用是造数字，算数据
# matplotlib.pyplot 的作用是画图表，数据可视化
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 生成 x 轴数据：从 -5 到 5 取 100 个点
x = np.linspace(-5,5,100)

# 计算 y 值
y = x**2

# 画图
plt.plot(x,y)

# 加标题和坐标值标签
plt.title("$y = x^2$")
plt.xlabel("x")
plt.ylabel("y")

# 显示网格
plt.grid(True,alpha=0.3)

# 展示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
x = np.linspace(-5,5,100)
# 继续用刚才的 x 数据
y = x**2
y2 = x**3
y3 = x**4

plt.figure(figsize=(10,6)) #设置画布的大小

plt.plot(x,y,label="$y = x^2$",linewidth=2)
plt.plot(x,y2,label="$y = x^3$",linewidth=2)
plt.plot(x,y3,label="$y = x^4$",linewidth=2)

plt.title("幂函数对比")
plt.xlabel("x")
plt.ylabel("y")
plt.legend() # 显示图例
plt.grid(True,alpha=0.3) # 半透明网格
plt.axhline(y=0,color='black',linewidth=0.5) # x轴
plt.axvline(x=0,color='black',linewidth=0.5) # y轴

plt.show()
```

## 观察笔记  
### 幂函数的共同点 - 所有幂函数都经过原点 $(0, 0)$。 - 当 $x > 0$ 时，所有函数值都是正数。  
### 幂函数的不同点 - 偶次幂（$x^2$、$x^4$）关于 $y$ 轴对称，图形在 $x$ 轴上方。 - 奇次幂（$x^3$）关于原点对称，图形穿过第三象限。 - 指数越大，在 $0 < x < 1$ 区间越平缓，在 $x > 1$ 区间上升越快。

### Summary
> Power functions grow at different rates depending on their exponent. 
> Even powers are symmetric about the y-axis, while odd powers are symmetric about the origin.
