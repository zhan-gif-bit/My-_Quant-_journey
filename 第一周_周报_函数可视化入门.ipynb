## 1. 导数：变化率的数学语言

导数的意义：
$$f'(x) = \lim_{\Delta x \to 0} \frac{f(x+\Delta x) - f(x)}{\Delta x}$$

### 基本求导公式
- 幂函数： $\frac{d}{dx} x^n = n x^(n - 1)$
- 正弦函数： $\frac{d}{dx} \sin x = \cos x$
- 指数函数:  $\frac{d}{dx} a^x = a^x lna$

> 导数描述了函数在某一点的**瞬时变化率**，是量化分析中理解变量的基石|

```python

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

x = np.linspace(-3,3,200)
y1 = x**2
y2 = x**3
y3 = x**4

plt.figure(figsize=(10,6))

plt.plot(x,y1,label="$y = x^2$",linewidth=2)
plt.plot(x,y2,label="$y = x^3$",linewidth=2)
plt.plot(x,y3,label="$y = x^4$",linewidth=2)

plt.title("幂函数对比图",fontsize=14)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True,alpha=0.3)
plt.axhline(y=0,color="black",linewidth=0.8)
plt.axvline(x=0,color="black",linewidth=0.8)


plt.show()
```
