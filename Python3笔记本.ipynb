# 导数的重要公式
反应函数在某一点的**瞬时变化率**

## 1. 导数的定义
设函数$y=f(x)$ 在点$x$处可导，则其导数定义为：
$$f'(x)=\lim_{\Delta x \to 0} \frac{f(x+\Delta x) - f(x)}{\Delta x}$$
这个极限如果存在，就是切线斜率

## 2. 基本求导公式
### 幂函数
$$\frac{d}{dx}（表明求导的专业符号） x^n=n x^{n-1}$$                                
或者(高中版)    
如果$$f(x)=x^n$$
则$$f'(x)=n x^{n-1}$$
### 三角函数
$$\frac{d}{dx} \sin x= \cos x$$
$$\frac{d}{dx} \cos x= -\sin x$$
### 指数函数与对数函数   
$$\frac{d}{dx} a^x=a^x lna$$
$$\frac{d}{dx} \log_{a} x=\frac{1}{x lna}$$

## 3. 四则运算求导法则

设 $u(x)$ 和 $v(x)$ 可导。

- **和差法则**：  
  $$\frac{d}{dx}(u \pm v) = \frac{du}{dx} \pm \frac{dv}{dx}$$

- **乘积法则**：  
  $$\frac{d}{dx}(uv) = u \frac{dv}{dx} + v \frac{du}{dx}$$

- **商法则**：  
  $$\frac{d}{dx}\left(\frac{u}{v}\right) = \frac{v \frac{du}{dx} - u \frac{dv}{dx}}{v^2}$$

> The derivative represents the instantaneous rate of change.     
> Mastering these basic rules is the first step towards understanding calculus.
