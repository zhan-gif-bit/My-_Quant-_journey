# Python数据结构入门

```python
# 创建一个列表
stocks = ["AAPL","GOOG","MSFT"]
print("股票列表",stocks)
print("第一个股票",stocks[0])
print("股票数量",len(stocks))

# 往列表里加东西
stocks.append("TSLA")

print("添加后",stocks)
print("新股票数量",len(stocks))
```

```python
# 用 for 循环遍历列表
prices = [150, 2800, 330, 250]

print("股票价格：")
for price in prices:
    print(f"${price}")
```

```python
# if 判断：找出涨跌幅最大的股票
change_rate = 1 # 单个等号是赋值

if change_rate > 0:
    print("上涨")
elif change_rate == 0:    #两个等号表示比较运算
    print("平盘")
else :
    print("下跌")
```

```python
# 创建一个字典，描述一只股票
stock_info = {                    
    "名称": "苹果",   # 注意字典里的每个元素要用逗号隔开
    "代码": "AAPL",
    "价格": "150.25",
    "涨跌幅":"2.3"
}

print("股票代码",stock_info["代码"])
print("涨跌幅",stock_info["涨跌幅"],"%")
```

```python
# 模拟三只股票的数据（字典列表）
portfolio = [
    {"代码":"APPL","价格":150,"涨跌幅":2.5},
    {"代码":"GOOG","价格":2800,"涨跌幅":-1.2},
    {"代码":"TSLA","价格":250,"涨跌幅":0.8},
]

# 筛选出上述股票
print("今日上涨的股票:")
for stock in portfolio: 
    if stock["涨跌幅"] > 0:
# :+ 自动根据数值添加正/负号
        print(f"{stock['代码']}: {stock['涨跌幅']:+}%")
```

```python
# f"···"用来拼接字符串和变量
name = "张三"
score = 95
print(f"{name}的分数是{score}分")
```
